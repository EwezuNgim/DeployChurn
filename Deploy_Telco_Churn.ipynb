```python
import pickle
import pandas as pd
from flask import request,jsonify,Flask
from waitress import serve
input_file="C://Users//FACULTY OFFICIAL//model_C=boost.bin"
with open(input_file,'rb') as f_in:
    dv,SC,boost=pickle.load(f_in)
app=Flask(__name__)
@app.route('/',methods=['POST'])
def predict():
    customer=request.get_json()
    if customer is None:
            return jsonify({"error": "Invalid JSON payload"}), 400
    customer=pd.DataFrame([customer])
    customer.iloc[:,16:19]=SC.transform(customer.iloc[:,16:19].values)
    customer=customer.to_dict(orient='records')
    X=dv.transform(customer)
    y_pred= boost.predict_proba(X)[0,1]
    churn=y_pred>=0.5
    result={
        'churn_probability':float(y_pred),
        'churn':bool(churn)
        }
    #float and bool to convert them from numpy.float64 and numpy.bool
    return jsonify(result)
if __name__ =='__main__':
    app.run(host='0.0.0.0',port=9696)
    #serve(app,host='0.0.0.0',port=8080)
```
